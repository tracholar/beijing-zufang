```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.style='ggplot'
%matplotlib inline

import re
import numpy as np
```

```python
df = pd.read_csv('fj_20160403.txt', sep='\t',names=['title','URL','INFO','ADDR','price'], header=False)
df.head()
```

```python
df['region'] = df.ADDR.map(lambda x: x[:x.find('-')])
```

```python
pt = re.compile(r'(\d+)㎡')
def get_area(x):
    res = pt.search(x)
    if res:
        return float(res.groups()[0])
    else:
        return np.inf
df['area'] = df.INFO.map(get_area)
df.area.head()
```
